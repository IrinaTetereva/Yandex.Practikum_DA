# Название проекта: Поиск факторов успешности игры 

**Описание проекта**
Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).  
У нас данные до 2016 года. Сейчас декабрь 2016 г., и мы планируем рекламную кампанию на 2017-й.  
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating
Board) — это ассоциация, определяющая возрастной рейтинг компьютерных
игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную
категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для
подростков».
  
**Цель**
  выявить закономерности, определяющие успешность игры. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
  

**Задачи и ход исследования**

1. Изучить общую информацию и подготовить данные  
  
*   Привести к нижнему регистру названия столбцов
*   Преобразовать данные в нужные типы. Описать, в каких столбцах заменили тип данных и почему
*   Обработать пропуски при необходимости
*   Посчитать суммарные продажи во всех регионах и запишите их в отдельный столбец

2. Провести исследовательский анализ данных
*  Посмотреть, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?  
*  Посмотреть, как менялись продажи по платформам. Выбрать платформы с наибольшими суммарными продажами и построить распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?
*  Взять данные за соответствующий актуальный период. Какие платформы лидируют по продажам, растут или падают? Выберать несколько потенциально прибыльных платформ.
*  Построить график «ящик с усами» по глобальным продажам игр в разбивке по платформам. Описать результат.
*  Построить диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулировать выводы.
*  Соотнести выводы с продажами игр на других платформах.
*  Посмотреть на общее распределение игр по жанрам. Что можно сказать о самых прибыльных жанрах? Выделяются ли жанры с высокими и низкими продажами?  
  
3. Составить портрет пользователя каждого региона (NA, EU, JP)  
 
  
4. Проверить и пояснить гипотезы  
*  Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
*  Средние пользовательские рейтинги жанров Action и Sports разные.
*  Как вы сформулировали нулевую и альтернативную гипотезы;
*  Какой критерий применили для проверки гипотез и почему.  

5. Написать общий вывод



**Описание данных**  
`Name` — название игры  
`Platform` — платформа  
`Year_of_Release` — год выпуска  
`Genre` — жанр игры

`JP_sales` — продажи в Японии (миллионы проданных копий)  
`Other_sales` — продажи в других странах (миллионы проданных копий)

`User_Score` — оценка пользователей (максимум 10)  
`Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating 
Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает
им подходящую возрастную категорию.  
Данные за 2016 год могут быть неполными.

## Преобразование данных в нужные типы.  
На первый взгляд, неправильные типы данных обнаружены в следующих колонках:  *year_of_release* и *user_score*:
- *year_of_release* имеют тип **float64** вместо **int** (лишние нули нам не нужны), в столбце должны хранится данные о годе выпуска, логично заменить тип данных с дробного на формат даты, но т.к. у нас только го, поэтому достаточно и int.  
- *user_score* имеет тип **object** вместо **float64**, колонка должна хранить данные о пользовательской оценке, которые измерены в десятичном формате,  значение оценки tbd - это To Be Determined, то есть "Будет определено", логично, если это заменим на пропуски (Nan), поэтому приведём эту колонку к типу  **float64**, а ошибки переведём в nan.

```python
# Заменим типы данных
df.year_of_release = df.year_of_release.astype('int', errors='ignore')
df ['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
df.info()
```

В этих колонках много пропусков. Удалять их не будем, т.к. колонки содержат вспомогательную информацию для проведения исследования. Nan в них нам не мешает.  
Посмотрим описательные статистики в каждой.

```python
for column in ['critic_score', 'user_score', 'rating']:
    display(df[column].describe())
    display(df[column].value_counts())
```

Распределение значений в колонке рейтинга неоднородное: значительно больше значений у Е, далее Т, М и Е10+. Посмотрим в википедии что означает каждое значение рейтинга (https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board).  
1) **«EC»** («Early childhood») — «Для детей младшего возраста».  
2) **«E»** («Everyone») — «Для всех». Первоначально "K-A" ("Kids to Adults")  
3) **«E10+»** («Everyone 10 and older») — «Для всех от 10 лет и старше».  
4) **«T»** («Teen») — «Подросткам».  
5) **«M»** («Mature») — «Для взрослых».  
6) **«AO»** («Adults Only 18+») — «Только для взрослых». Рейтинг «Только для взрослых» является предметом многочисленных дискуссий, так как накладывает серьёзные ограничения на продажи игры.  
7) **«RP»** («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.  
  
Из описания видно, что значение "K-A" - это то же самое, что и Е. Произведём замену.

```python
df.loc[(df['rating']=='K-A'),'rating'] = 'E'
display(df['rating'].value_counts()) # Проверка замены
```

```python
# Найдём лидеров по продажам, т.е. те платформы, которые показали наибольшие суммарные продажи за 4 года
top_platforms_after_2013 = df.query('year_of_release>=2013').groupby(['platform',
                                                                      'year_of_release']).agg(
                                                                                              sales=('total_sales','sum')
                                                                                              ).reset_index()
top_platforms_after_2013 = top_platforms_after_2013.groupby('platform').agg(
                                                                            sum_sales=('sales','sum')
                                                                            ).sort_values(by='sum_sales', 
                                                                                          ascending=False).reset_index()
top_platforms_after_2013.head() 
```

2-3) Найдём растущие и падающие платформы

Растущая - это такая платформа, объём продаж которой с течением времени растёт. Т.к. данные у нас за 2016 год неполные, мы не можем адекватно судить о продажах за этот период. Посмотрим, какие платформы показали большие продажи в 2015 году, чем 2014, падающие - соответственно, те, которые показали отрицательный рост, в 2015 году их продажи ниже, чем в 2014.

```python
growing_platforms = df.query('year_of_release>2013 and year_of_release<2016').pivot_table(index = 'platform',
                                              columns = 'year_of_release',
                                              values = 'total_sales',
                                              aggfunc = 'sum',
                                              fill_value=0).reset_index()
def growth(row):
    result = row[2015] / row[2014]
    if result >= 1:
        return True
    else: return False

growing_platforms['is_growing_platform'] = growing_platforms.apply(growth, axis=1)
display(growing_platforms)
list_of_growing_platforms = growing_platforms.query('is_growing_platform==True')['platform'].to_list()
display('Растущими на начало 2016 года были:', list_of_growing_platforms)
display('Падающими на начало 2016 года были:', growing_platforms.query('is_growing_platform==False')['platform'].to_list())
```

### Вывод
В 2016 году в живых остаются следующие платформы: PS4, PS3, XOne, 3DS, X360, причём явно растущими являются платформы PS4 и XOne. Остальные показывают отрицательную динамику продаж.

## Проверка гипотез
### Проверим двустороннюю гипотезу о равенстве средних оценок пользователей платформ Xbox One и PC 

$\begin{equation*}
 \begin{cases}
   H_0 :\text{Средние пользовательские оценки платформ Xbox One и PC одинаковые}\\
   H_1 :\text{Средние пользовательские оценки платформ Xbox One и PC различаются}
 \end{cases}
\end{equation*}$

Посмотрим на выбросы в значениях оценок пользователей для платформ Xbox One и PC: 

```python
for col in ["XOne","PC"]:
    df_2013.query('platform == @col').boxplot('user_score')
    plt.title(f'{col}')
    plt.show()
```

```python
for col in ["XOne","PC"]:
    q1_limit_low_user_score = df_2013.query('platform == @col')['user_score'].quantile(0.25)
    q3_limit_low_user_score = df_2013.query('platform == @col')['user_score'].quantile(0.75)
    iqr_user_score = q3_limit_low_user_score - q1_limit_low_user_score # iqr
    
# границы нижнего интервала наиболее вероятных значений
    limit_low_user_score = q1_limit_low_user_score - 1.5 * iqr_user_score
    print(f'границы нижнего интервала наиболее вероятных значений {col}:', round(limit_low_user_score, 2))
```

Значения меньшие 3.25 для XOne и 1,85 для PC - выбросы, удалим их, т.к. тест Стьюдента на проверку гипотез чувствителен к выбросам

```python
df_2013_XOne = df_2013.query('platform == "XOne"')[df_2013.query('platform == "XOne"')['user_score'] > 3.25]
df_2013_PC = df_2013.query('platform == "PC"')[df_2013.query('platform == "PC"')['user_score'] > 1.85]

print(df_2013_XOne.shape)
print(df_2013_PC.shape)
```

Значений достаточно для проверки гипотез по тесту Стьюдента.  
  
Чтобы проверить гипотезы о равенстве средних оценок пользователей платформ Xbox One и PC необходимо:  
1) установить, есть ли зависимость между двумя генеральными совокупностями;  
2) посмотреть распределение выборочных средних для двух совокупностей;  
3) доказать/опровергнуть равенство дисперсий двух генеральных совокупностей

Пороговое значение альфа зададим равное 0,01    
  
Итак, (1) в нашем случае мы рассматриваем генеральные совокупности пользователей двух разных платформ, они друг с другом **не связаны**, т.к. представлены людьми с разными потребностями. (2) В нашей выборке всего 180+149=329 человек, поэтому, ссылаясь на ЦПТ, можно утверждать, что распределение выборочных средних для двух совокупностей будет **нормальным**. (3) Дисперсии двух генеральных совокупностей скорее всего не равны, т.к. рассматриваемые платформы предполагают разное игровое поведение пользователей.  
  
Следовательно, для проверки гипотезы о равенстве средних применим ttest_ind из библиотеки scipy.stats, параметр equal_var зададим False.  
Найдём p-value. Если p-value будет меньше выбранного уровня значимости  𝛼=0.01 , то нулевую гипотезу о равенстве средних отвергнем. Иначе сделаем вывод, что данные не позволили нам отвергнуть нулевую гипотезу.

```python
alpha = 0.01
results = st.ttest_ind(df_2013_XOne['user_score'], df_2013_PC['user_score'], equal_var=False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Вывод: Отвергаем нулевую гипотезу")
else:
    print("Вывод: Не получилось отвергнуть нулевую гипотезу") 
```

#### Вывод
Результат проверки по тесту Стьюдента показал, что не получилось отвергнуть нулевую гипотезу, значит, средние оценки пользователей платформ Xbox One и PC могут не различатся.

### Проверим двустороннюю гипотезу о равенстве средних оценок пользователей  для жанров Action и Sports

$\begin{equation*}
 \begin{cases}
   H_0 :\text{Средние пользовательские оценки жанров Action и Sports одинаковые}\\
   H_1 :\text{Средние пользовательские оценки жанров Action и Sports различаются}
 \end{cases}
\end{equation*}$

Посмотрим на выбросы в значениях оценок пользователей для жанров Action и Sports :

```python
df_2013.query('genre in ["Action","Sports"]').boxplot('user_score');
```

```python
q1_limit_low_user_score_genres = df_2013.query('genre in ["Action","Sports"]')['user_score'].quantile(0.25)
q3_limit_low_user_score_genres = df_2013.query('genre in ["Action","Sports"]')['user_score'].quantile(0.75)
iqr_user_score_genres = q3_limit_low_user_score_genres - q1_limit_low_user_score_genres # iqr
    
# границы нижнего интервала наиболее вероятных значений
limit_low_user_score_genres = q1_limit_low_user_score_genres - 1.5 * iqr_user_score_genres
limit_low_user_score_genres
```

Значения меньшие 2,35 - выбросы, удалим их, т.к. тест Стьюдента на проверку гипотез чувствителен к выбросам

```python
df_2013_Action = df_2013.query('genre == "Action"')[df_2013.query('genre == "Action"')['user_score'] > limit_low_user_score_genres]
df_2013_Sports = df_2013.query('genre == "Sports"')[df_2013.query('genre == "Sports"')['user_score'] > limit_low_user_score_genres]

print(df_2013_Action.shape)
print(df_2013_Sports.shape)
```

Рассуждения аналогичные проверке предыдущих гипотез.  
Пороговое значение альфа зададим равное 0,01    
  
Итак, (1) генеральные совокупности пользователей **не связаны**. (2) Ссылаясь на ЦПТ, можно утверждать, что распределение будет **нормальным**. (3) Дисперсии двух генеральных совокупностей **не равны**.  
  
Следовательно, для проверки гипотезы о равенстве средних снова применим ttest_ind из библиотеки scipy.stats, параметр equal_var зададим False.  
Найдём p-value. Если p-value будет меньше выбранного уровня значимости  𝛼=0.01 , то нулевую гипотезу о равенстве средних отвергнем. Иначе сделаем вывод, что данные не позволили нам отвергнуть нулевую гипотезу.

```python
alpha = 0.01
results = st.ttest_ind(df_2013_Action['user_score'], df_2013_Sports['user_score'], equal_var=False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Вывод: Отвергаем нулевую гипотезу")
else:
    print("Вывод: Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод
Результат проверки по тесту Стьюдента показал, что необходимо отвергнуть нулевую гипотезу, значит, средние оценки пользователей игр жанров Action и Sports различаются. Т.к. тест двухсторонний, то мы не знаем точно оценки игр каких жанров выше в среднем.
